### ***Relationship between X and Y***

X     Y  
2     3  
4     5  
6     7  
8     9  

---

#### ***Direction based Relationship***

POSITIVE RELATIONSHIP  

X ↑     Y ↑  
X ↓     Y ↓  

NEGATIVE RELATIONSHIP  

X ↑     Y ↓  
X ↓     Y ↑  

---

#### ***Scatter Plot Interpretation***

POSITIVE CORRELATION  
• As X increases, Y also increases  
• Points move upward  

NEGATIVE CORRELATION  
• As X increases, Y decreases  
• Points move downward  

---

#### ***Covariance***

Cov(X , Y) =  
Σ ( xi − x̄ )( yi − ȳ )  / 
        n − 1  

Where,  
xi  → datapoint of X  
x̄   → sample mean of X  
yi  → datapoint of Y  
ȳ   → sample mean of Y  

---

#### ***Variance***

Var(X) =  
Σ ( xi − x̄ )²  /  
   n − 1  

Var(X) = Cov(X , X)  

→ Variance shows SPREAD of data  

---

#### ***Covariance Sign***

POSITIVE COVARIANCE  

X ↑     Y ↑  
X ↓     Y ↓  

NEGATIVE COVARIANCE  

X ↑     Y ↓  
X ↓     Y ↑  

---

#### ***Example***

X     Y  
2     3  
4     5  
6     7  

x̄ = 4  
ȳ = 5  

Cov(X , Y) =  
(2−4)(3−5) + (4−4)(5−5) + (6−4)(7−5)  /  
                 2  

= (4 + 0 + 4) / 2  
= 8 / 2  
= 4  

→ Positive Covariance  

---

####  ***Advantages***

1. Shows relationship between X and Y  
2. Shows +ve or −ve direction  

---

#### ***Disadvantages***

1. No fixed limit value  
2. Hard to interpret  

---

#### ***Pearson Correlation Coefficient***

Range :  −1  to  +1  

r(x , y) =  
Cov(X , Y)  /
 σx   σy  

---

#### ***Interpretation***

r → +1  : Strong Positive Correlation  
r → −1  : Strong Negative Correlation  
r = 0   : No Linear Relationship


### (3) Spearman Rank Correlation   [ -1 to 1 ]
