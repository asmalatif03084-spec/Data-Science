### ***GENERAL USE OF EXCEPTION***
##### ***ZerodiVisionError***

```python
try:
    a=10/0
except ZeroDivisionError as e:
    print(e)    
```

##### ***ValueError&TypeError***

```python
try:
    int("asma")
except (ValueError,TypeError) as e:
    print(e)    
```

##### ***Except***

```python
try:
    int("asma")
except:
    print("this will catch an error")    
```

##### ***ImportError***

```python
try:
    import asma
except ImportError as e:
    print(e) 
10/4   
```

##### ***KeyError***

```python
try:
    d={"key":"value",1:[1,2,3,4]}
    print(d["key2"])
except KeyError as e:
    print(e)    
```

##### ***AttributeError***

```python
try:
    "Asma".test()
except AttributeError as e:
    print(e)    
```

##### ***IndexError***

```python
try:
    l=[1,2,3,4]
    print(l[6])
except IndexError as e:
    print(e)    
```

##### ***TypeError***

```python
try:
    123+"asma"
except TypeError as e:
    print(e)    
```

##### ***FileNotFoundError***

```python
try:
    with open("test5.txt",'r') as f:
        test=f.read()
except FileNotFoundError as e:
    print(e)        
```

##### ***Use Of SuperClass***

```python
try:
    with open("test5.txt",'r') as f:
        test=f.read()
except Exception as e:
    print(e)        
except FileNotFoundError as e:
    print("asma",e)  
```

```python
def test(file):
    try:
        with open(file,'r') as f:
            test=f.read()
    except  Exception as e:
        print(e)        
    except FileNotFoundError as e:
         print("asma",e)  
```
