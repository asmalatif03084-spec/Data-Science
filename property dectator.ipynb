### ***PROPERTY DECTATOR***
#### ***SETTER,GETTER,DELETE***

```python
class asma:
    def __init__(self,account_name,account_number):
        self.account_name=account_name
        self.__account_number=account_number

#GETTER
    @property
    def account_number_access(self):
        return self.__account_number  
#SETTER
    @account_number_access.setter
    def account_number_set(self,price):
        if price<=3500:
            pass
        else:
            self.__account_number=price

#DELETE
    @account_number_access.deleter
    def delete_account_number(self):
        del self.__account_number        
```

```python
ab=asma("HBL",23456)
```
