## ***LIST***

```python
l=["asma",1,2,3,4,33.24,True]
print(l[4])
print(l)
```

## ***slicing in a list***

```python
print(l[0:3]) #list ki range print krwani hu
print(l[-1]) #list ka last element print krwana hu
print(l[::-1]) #list ko reverse order mn print krwana hu tu
print(l[::2]) #print only that are at the difference of 2
print(l[5]) #specific word print krana hu list ka
print(l[0][0:2])#agr list ky  ik index ky word ky andr slicing krni hu
l1=[13,14,15]
print(l+l1)
```

## ***INSERTION***
### ***append,extend,insert***

```python
l2=[1,2,3,4,5,6]
l2.append("asma")
print(l2)
l2.extend("saad")
print(l2)
l2.insert(2,"asma")
print(l2)
```

## ***DELETION***
### ***pop,remove,sort,reverse***


```python
l3=[1,2,3,4,5,6,7]
l3.pop(3)#index ky through value delete huti ha
print(l3)
l3.remove(5)#value ky through delete huti ha
print(l3)
l4=[1,2,3,4,[1,2,4,5,6]]
l4[4].remove(1)
print(l4)
l5=[10,15,3,4,9,0,23,45,555,666]
l5.sort()
print(l5)
l5.sort(reverse=True)
print(l5)
l5.reverse()
print(l5)
```
