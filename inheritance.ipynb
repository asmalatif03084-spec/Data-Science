### ***INHERITANCE***

```python
class test:
    def method(self):
        return "this is my first class of test"
```

```python
class child(test):
    pass
```

```python
asma=child()
```

```python
asma.method()
```

```python
class class1:
    def test_class1(self):
        return "this is my first class1"
```

```python
class class2(class1):
    def test_class2(self):
        return "this is my second class2"
```

```python
class class3(class2):
    pass
```

```python
asma=class3()
```

```python
asma.test_class1()
```

```python
class class1:
    def test_class1(self):
        return "this is my first class1"
```

```python
class class2():
    def test_class2(self):
        return "this is my second class2"
```

```python
class class3(class1,class2):
    pass
```

```python
asma=class3()
```

```python
asma.test_class1()
```
