```python
l=[1,2,3,4,"asma","latif",[1,2,3,4]]
def test7(l):
    l1=[]
    for i in l:
        if type(i) == int or type(i) == float:
            l1.append(i)
    return l1
test7(l)        
        
```

```python
a=[1,2,3,4,"asma","latif",[1,2,3,4]]
def test8(a):
    l=[]
    for i in a:
        if type(i) ==list:
            for j in i:
                l.append(j)
        else:
                if type(i)==int or type(i)==float:
                    l.append(i)
    return l
                       
        
```

```python
print(test8(a))
```

### ***comprehenshion***

#### ***list comprehenshion***

```python
l=[1,2,3,4,5]
l1=[]
for i in l:
    l1.append(i**2)
```

```python
l1
```

```python
#comprehensive
l=[1,2,3,4,5]
[i**2 for i in l]
```

```python
l=[1,2,3,4,5,6]
[i for i in l if i%2==0 ]
```

```python
l1=["asma","saba","saad","umaira"]
[i.upper() for i in l1]
```

#### ***tuple comprehension***

```python
l=[1,2,3,4,5]
list(i**2 for i in l)
```

#### ***dictionary comprehension**

```python
d={"key1":1,"key2":2,"key3":3,"key4":4}
{k:v**2 for k,v in d.items()}
```

```python
d={"key1":1,"key2":2,"key3":3,"key4":4}
{k:v for k,v in d.items() if v > 1}
```
