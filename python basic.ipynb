## ***INDEXING***

```python
a="asma"
print(a[0]) #from start
print(a[-1]) #from end
print(a[0:3]) #slicing
print(a[0:4:2]) #jumping in slicing 0:4 indicate total length of name 2 indicate no of jumping
len(a) #indicate total length
```

```python
a.find("m") #find tell us the index no 
a.upper() #show  all uppercase
a.lower() #show all smaller
a.count("a") #it shows how many time a occur in our string
a.capitalize() #capitalize first letter
```
