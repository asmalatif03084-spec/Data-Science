```python
class car:
    def __init__(self,year,model,make,speed):
        self.__year=year
        self.__model=model
        self.__make=make
        self.__speed=speed

    def set_speed(self,speed):
        self.__speed= 0 if speed<0 else speed

    def get_speed(self):
        return self.__speed        
        
```

```python
a=car(2025,"innovo","landcruser",12)
```

```python
a.get_speed()
```

```python
class bankaccount:
    def __init__(self,balance):
        self.__balance=balance


    def deposit(self,amount):
        self.__balance=self.__balance+amount

    def withdraw(self,amount):
        if self.__balance>=amount:
            self.__balance=self.__balance-amount
            return True
        else:
            return False


    def get_balance(self):
        return self.__balance            
    
```

```python
asma=bankaccount(1000)
```

```python
asma.get_balance()
```

```python
asma.deposit(1900)
```

```python
asma.get_balance()
```

```python
asma.withdraw(3000)
```

```python
asma.withdraw(2500)
```

```python
asma.get_balance()
```
