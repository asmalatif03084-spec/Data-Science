### ***TYPES OF DATA***

#### ***There are two types of data.***
#### ***1:Quantative***
#### ***2:Qualitatitve***

##### **1: Quantative:It is a numerical data on which we can perform mathematical task.It further divided into two types.**
#### **Discrete:It give whole value.For example:No of childern,No of bank account.**
#### **Continous:It give any type of value.For example:Height,Weight,Speed.**

#### ***2: Qualitative:It is a categorical data.it further divided into two types.***
##### ***Nominal:It does not give any rank.For example:Gender,Blood Group***
##### ***Ordinal:It give rank.For example:Coustomer Feedback Good,Average,Bad***

### ***LEVEL OF MEASUREMENT***

#### ***1:Nominal Scale Data***
##### ***Categorical/Quantative***
##### ***E.g:Gender,Colors***
##### ***Order does not matter.***

#### ***2: Ordinal Scale Data***
##### ***It give rank.***
##### ***Order matter.***
##### ***Difference can not measured.***
##### ***Example:Feedback from coustomer.***

#### ***3: Interval Scale Data***
##### ***Ratio can not be measured.***
##### ***Order matter.***
##### ***Difference can be measured.***
##### ***No "0" starting point.***
##### ***Example:Temperature***

#### ***4: Ratio scale Data***
##### ***Ratio are measured.***
##### ***Order matter.***
##### ***Difference are measured.***
##### ***Contain a "0" starting point.***
##### ***Example:Marks of student***
