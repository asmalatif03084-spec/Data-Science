### ***FETCHING DATA***
#### ***index concept not applicable in dict we use key***

```python
d5={'key':'asma',123:'84','-war':'saba'}
d5[123]
d5['-war']
```

```python
d5={'key':'asma',123:'84','-war':'saba','key':'asmalatif'}#key unique na hu tu overwrite huta ha new sy
d5['key']#dictionary mn agr duplicate key hu tu jo last mn key hu wo execute hugi
```

```python
d5={'name':'asma','course':["web","flutter","python","data science"]}
print(d5)
d5['course']#list use huskti ha dictionary mn
d5['course'][2]#hm list mn sy specific data bi print krwa skty hain
```

```python
#as a value we can also take list,tuple or set 
#also use dictionary inside a dictionary
d6={"numbers":[1,2,3],"assignment":(1,2,3),"dates":{23,14},"time":{"web":"12","dive":"1"}}
print(d6)
d6['time']['dive']#dictionary ky andr dict ka data fetch krna
```

```python
#add new key in a dict
d6={'name':'asma','roll-no':'84'}
d6['qualification']="graduate"
print(d6)
```

```python
#key available hu tu update krde ga
d6={'name':'asma','roll-no':'84'}
d6["name"]="sadia"
print(d6)
```

```python
#deletion
d6={'name':'asma','roll-no':'84'}
del d6["name"]
print(d6)
```
