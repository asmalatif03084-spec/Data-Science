### ***STATISTICS AND TYPES OF STATISTICS***

##### ***Statistic is the science of collecting,organizing and analyzing data.its primary purpose is to help in decision making based on evidence***
##### ***Data: Defined as "Facts or piece of information***

#### ***Types of Statistics***
#### ***Descriptive statistic:***

##### ***Descriptive statistic consist of organizing and summarizing data***
##### ***Measure of centeral Tendency:Mean,Mode,Median***
##### ***Measure of Dispersion:Vatiance,Standard deviation***

#### ***Inferential Statistic:***

##### ***Inferential statistic consist of using the data you have measured to form conclusion***
##### ***Z-test,T-test,chi square test***


#### ***Population and Sample Data***

##### ***Population:The entire group you are interested in studying.***
##### ***Sample:A subset of the population.***


#### ***EXAMPLE***
##### ***There are 20 statistic classes at your college.you collect the heights of students in specific class.Recorder Height:120cm,160cm,150cm.***
##### ***Descriptive : What is average height of student?***
##### ***Inferential: Are the height of students in this classroom similar to what you wouldexpect for the entire college?***
