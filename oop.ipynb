```python
class person:
    def __init__(self,Name,Age,Group):
        
        self.name=Name
        self.age=Age
        self.group=Group
        print(self.name,self.age,self.group)
    def return_detail(self):
        return self.name,self.age,self.group
             
       
```

```python
a=person("asma",12,"b")
```

```python
a.return_detail()
```

```python
b=person("sadia",25,"c")
```

```python
b.return_detail()
```
