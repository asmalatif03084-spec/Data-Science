```python
class data_science:
    def syllabus(self):
        print("this is my syllabus for data science")
```

```python
class dev:
    def syllabus(self):
        print("this is my syllabus fr dev")
```

```python
def class_pref(class_obj):
    for i in class_obj:
        i.syllabus()
```

```python
data_Science=data_science()
```

```python
dev=dev()
```

```python
class_obj=[data_Science,dev]
```

```python
class_pref(class_obj)
```
