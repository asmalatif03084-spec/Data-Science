### ***EXCEPTION HANDLING***
#### ***It consist four block Try,Except,Else,Finally***
##### ***Try execute when there is no error except execute when try block not successfully execute***

```python
try:
    f=open("test.txt","r")
except Exception as e:
    print("this is except error",e)    
```

```python
try:
    f=open("test.txt","w")
    f.write("write this in my file")
    f.close()
except Exception as e:
    print("this is except error",e)  
```

##### ***Else execute when try execute successfully***

```python
try:
    f=open("test.txt","w")
    f.write("write this in my file")
    # f.close()
except Exception as e:
    print("this is except error",e) 
else:
    f.close()   
    print("this block will execute when your try block succcessfully execute") 
```

```python
try:
    f=open("test1.txt","r")
    f.write("write this in my file")
    # f.close()
except Exception as e:
    print("this is except error",e) 
else:
    f.close()   
    print("this block will execute when your try block succcessfully execute") 
```

##### ***Finally execute in any situation***

```python
try:
    f=open("test2.txt","w")
    f.write("this is my file")
finally:
    print("finally will always execute in any situation")    
```
