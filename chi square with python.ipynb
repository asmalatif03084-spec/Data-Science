### ***CHI SQUARE WITH PYTHON***

```python
import scipy.stats as stat
import numpy as np
```

```python
#no of student studied in a day weekly basic
expected_hour=[8,6,7,9,6,9,7]
observed_hour=[7,8,6,9,9,6,7]
```

```python
#chi square goodness of fit
chi_square_statistic,p_value=stat.chisquare(observed_hour,expected_hour)
```

```python
chi_square_statistic,p_value
```

```python
#find the critical value
significance_value=0.05
dof=len(expected_hour)-1
print(dof)
critical_value=stat.chi2.ppf(1-significance_value,dof)
```

```python
critical_value
```

```python
if chi_square_statistic > critical_value:
    print("reject the null hypothesis")
else:
    print("accept null hypothesis")    


```
