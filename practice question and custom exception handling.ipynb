#### ***Always use proper exception***

```python
try:
    a=10/0
except ZeroDivisionError as e:
    print(e)    
```

#### ***Always print a proper message in except***

```python
try:
    a=10/0
except ZeroDivisionError as e:
    print("i am trying to handel a zero division error",e)    
```

#### ***Always try to log your error***

```python
import logging
logging.basicConfig(filename="error.log",level=logging.ERROR)
try:
    a=10/0
except ZeroDivisionError as e:
    logging.error("i am trying to handel a zero division error{}".format(e))    
```

#### ***Always avoid to write a multiple exception handling***
##### ***using multiple exception only increse complexity***

```python
import logging
logging.basicConfig(filename="error.log",level=logging.ERROR)
try:
    a=10/0
except AttributeError as e:
    logging.error("i am trying to handel attribute error {}".format)   
except KeyError as e:
    logging.error("i am trying to handel  keyerror {}".format)       
except ZeroDivisionError as e:
    logging.error("i am trying to handel a zero division error{}".format(e))    
```

### ***CUSTOM EXCEPTION HANDLING***

```python
class validage(Exception):
    def __init__(self , msg):
     self.msg=msg
```

```python
def validage(age):
    if age<0:
        raise validage('enter age is negative')
    elif age>200:
        raise validage("enter age is very high")
    else:
        print("age is valid")
```

```python
try:
    age=int(input("Enter your age"))
    validage(age)
except validage as e:
    print(e)    
```
