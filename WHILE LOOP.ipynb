### ***WHILE LOOP***

```python
a=1
while a<=10:
    print(a)
    a=a+1
```

#### ***PRINT SUM OF NUMBER TILL A NUMBER***

```python
number=int(input("enter a number"))
start=0
counter=1
while counter<=number:
    start=start+counter
    counter=counter+1
    print(start)
```

#### ***CALCULATE FACTORIAL***

```python
n=int(input("enter a number"))
fact=1
num=1
while num<=n:
    fact=fact*num
    num=num+1
    print(fact)
```

#### ***FINNOIC SERIES***

```python
n=int(input("enter a number"))
n1=0
n2=1
counter=0
while counter<=n:
    print(n1)
    c=n1+n2
    n1=n2
    n2=c
    counter=counter+1
```

#### ***PRINT TABLE***

```python
n=int(input("enter a number"))
i=1
while i<=10:
    print( i,"*",i*n)
    i=i+1
```

#### ***REVERSE A STRING***

```python
word=str(input("enter a word"))
reverse=""
length=len(word)
while length>0:
    reverse=reverse+word[length-1]
    length=length-1
    print(reverse)
```
