### ***CHI SQUARE TEST***

The chi square test for goodness of fit test claims about population proportion.[categorical variable]

It is a non parametric test perform on categories [ nominal,ordinal] data.

For example:There is population of male who like different color of bike.

Yellow bike 1/3            22
 
orange bike 1/3            17

Red bike    1/3            59

### ***Goodness of Fit Test***

**In a student class of 100 students, 30 are right handed.  
Does this class fit the theory that 12% of people are right handed?**



####  ***Given***
- Total students = 100  
- Observed right handed students = 30  
- Theory: 12% people are right handed  





#### ***Observed (O) and Expected (E) Frequencies***

| Category       | Observed (O) | Expected (E) |
|---------------|-------------|--------------|
| Right handed  | 30          | 12           |
| Left handed   | 70          | 88           |
| **Total**     | **100**     | **100**      |



#### ***Notes***
- Theory gives a **categorical distribution**
- Observed values come from the **sample**


#### ***Chi-Square Goodness of Fit Test***

#### ***Given (2010 data)***


| Weight (kg) | Percentage |
|------------|------------|
| < 50       | 20%        |
| 50 – 75    | 30%        |
| > 75       | 50%        |



#### ***Observed Data (2020)***

Sample size = 500

| Weight (kg) | Observed (O) |
|------------|--------------|
| < 50       | 140          |
| 50 – 75    | 160          |
| > 75       | 200          |


#### ***Expected Frequencies***

| Weight (kg) | Expected (E) |
|------------|--------------|
| < 50       | 100          |
| 50 – 75    | 150          |
| > 75       | 250          |



#### ***Hypotheses***

- **H₀:** Data meets the expectation  
- **H₁:** Data does not meet the expectation  



#### ***Significance Level***

\[
\alpha = 0.05
\]



#### ***Degree of Freedom***

\[
df = k - 1 = 3 - 1 = 2
\]



#### ***Chi-Square Formula***

\[
\chi^2 = \sum \frac{(O - E)^2}{E}
\]

\[
\chi^2 =
\frac{(140-100)^2}{100}
+ \frac{(160-150)^2}{150}
+ \frac{(200-250)^2}{250}
\]

\[
\chi^2 = 16 + 0.66 + 10 = 26.66
\]



#### ***Decision Rule***

Critical value at \(df=2,\ \alpha=0.05\) is **5.99**

- If \(\chi^2 > 5.99\) → Reject \(H_0\)



#### ***Conclusion***

\[
\chi^2 = 26.66 > 5.99
\]

**Reject the Null Hypothesis**

**The weights of the 2020 population are different from those of 2010.**
