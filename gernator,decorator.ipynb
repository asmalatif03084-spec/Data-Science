#### ***generator function***

```python
range(0,10)
```

```python
for i in range(10):
    print(i)
```

```python
def fib(n):
    a,b=0,1
    for i in range(n):
        yield a
        a,b=b,a+b
```

```python
fib(10)
```

```python
for i in fib(10):
    print(i)
```

```python
def fib1():
    a,b=0,1
    while True:
        yield a
        a,b=b,a+b
```

```python
fib=fib1()
```

```python
for i in range(10):
    print(next(fib))
```

```python
def count_test(n):
    count=1
    while count<=n:
        yield count
        count=count+1
```

```python
c=count_test(5)
```

```python
for i in c:
    print(i)
```

### ***DECORATOR***

```python
def test():
    print("this is my function")
```

```python
test()
```

```python
def deco(func):
    def inner_dec():
        print("this is start")
        func()
        print("this is end")
    return inner_dec    
```

```python
@deco
def test1():
    print(6+7)
```

```python
test1()
```

```python
import time

def timer_test(func):
    def timer_test_inner():
        start=time.time()
        func
        end=time.time()
        print(end-start)
    return timer_test_inner
```

```python
@timer_test
def test2():
    print(9+10)
```

```python
test2()
```

```python
@timer_test
def test5():
    for i in range(10000000):
        pass
```

```python
test5()
```
