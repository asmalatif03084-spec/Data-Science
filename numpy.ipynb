#### **DATA TYPES**
#### **Int32,Int64,Int8,Float32,Float64,bool**

```python
a=np.array([1,2,3],np.int8)
b=np.array([1.5,2.8,3.9],np.float32)
c=np.array([True,False],np.bool)
print(a.dtype)
print(b.dtype)
print(c.dtype)
```

```python
print(a.itemsize)
print(b.itemsize)
print(c.itemsize)
```
