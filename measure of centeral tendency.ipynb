### ***2: MEDIAN***

#### ***Sort the random variable {1,2,2,3,4,5}***

No of element = 6

if count is even  

{1,2,2,3,4,5}

median=2+3/2=2.5                                

if count is odd

{1,2,2,3,4,5,6}

median=center element=3

#### ***Why we use median?***

{1,2,3,4,5}

mean=15/5=3

if we have outlier{1,2,3,4,5,100}

mean=115/6=19.2

while in median 4+3/2=3.5

Median is used  to find the centeral tendency when outlier is present.
